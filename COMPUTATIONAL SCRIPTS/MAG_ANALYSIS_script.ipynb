```r

# Create a summary data frame with counts
summary_data <- MAG_metadata %>%
  group_by(Collection.phase, Phylum, Class, Species.novelty, Strain.novelty) %>%
  summarize(count = n(), .groups = "drop")

# Order the levels of Species.novelty and Collection.phase
summary_data$Species.novelty <- factor(summary_data$Species.novelty, levels = c("yes", "no"))
summary_data$Collection.phase <- factor(summary_data$Collection.phase, levels = unique(summary_data$Collection.phase)[order(summary_data$Species.novelty)])

# Create the alluvial flow diagram
alluvial_plot <- ggplot(summary_data, aes(axis1 = Collection.phase, axis2 = Phylum, axis3 = Class, weight = count)) +
  geom_alluvium(aes(fill = Species.novelty), alpha = 0.8, width = 0.5, knot.pos = 0.5) +
  scale_fill_manual(values = c("yes" = "#FF7F0E", "no" = "#1F77B4")) +
  geom_stratum(width = 1/8, fill = "gray", color = "black") +
  geom_text(stat = "stratum", aes(label = after_stat(stratum)), size = 4, family = "Arial", fontface = "bold", color = "black") +
  theme_minimal() +
  labs(x = "Collection Phase", y = "Taxonomic Level", title = "Proportion of Species Novelty by Collection Phase and Taxonomic Level") +
  theme(plot.title = element_text(size = 12, face = "bold", hjust = 0.5, margin = margin(b = 10)),
        axis.text = element_text(size = 10, color = "black"),
        axis.title = element_text(size = 10, color = "black"),
        legend.title = element_blank(),
        legend.position = "bottom",
        legend.direction = "horizontal",
        legend.text = element_text(size = 10),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        plot.margin = margin(20, 20, 20, 20))

# Display the alluvial plot
alluvial_plot
```

```r
# Create a new variable to combine Species.novelty and Collection.phase
summary_data$Combined <- paste(summary_data$Species.novelty, summary_data$Collection.phase, sep = "-")

# Order the levels of the Combined variable
summary_data$Combined <- factor(summary_data$Combined, levels = c("no-Workday start", "yes-Workday start", "no-Workday end", "yes-Workday end", "no-Post-shower", "yes-Post-shower", "no-Swine", "yes-Swine", "no-Environment", "yes-Environment"))

# Reorder the Phylum levels by sum of count values
summary_data$Phylum <- fct_reorder(summary_data$Phylum, summary_data$count, sum)

# Create the alluvial flow diagram
alluvial_plot <- ggplot(summary_data, aes(axis1 = Collection.phase, axis2 = Phylum, weight = count)) +
  geom_alluvium(aes(fill = Species.novelty), alpha = 0.5, width = 0.5, knot.pos = 0.2) +
  scale_fill_brewer(type = "qual", palette = "Set1") +
  geom_stratum(width = 1/12, fill = "black", color = "grey") +
  geom_label(stat = "stratum", aes(label = after_stat(stratum))) +  
  theme_minimal() +
  labs(x = "Collection Phase", y = "Taxonomic Level", title = "Proportion of Species Novelty by Collection Phase and Taxonomic Level") +
  theme(plot.title = element_text(size = 12, face = "bold", hjust = 0.5, margin = margin(b = 10)),
        axis.text = element_text(size = 10, color = "black"),
        axis.title = element_text(size = 10, color = "black"),
        legend.title = element_blank(),
        legend.position = "bottom",
        legend.direction = "horizontal",
        legend.text = element_text(size = 10),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        plot.margin = margin(20, 20, 20, 20))

# Display the alluvial plot
alluvial_plot
```
